```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# List of boroughs in London excluding "City of London"
# List only contains first word of borough
# LSOA names start with the name of the borough the LSOA is in
boroughs = ["Westminster", "Kensington", "Hammersmith", "Wandsworth", "Lambeth", "Southwark", "Tower", "Hackney",
            "Islington", "Camden", "Brent", "Ealing", "Hounslow", "Richmond", "Kingston", "Merton", "Sutton", "Croydon",
            "Bromley", "Lewisham", "Greenwich", "Bexley", "Havering", "Barking", "Redbridge", "Newham", "Waltham", "Haringey",
            "Enfield", "Barnet", "Harrow", "Hillingdon"]
```

```python
df = pd.read_csv('imd2019lsoa.csv')
df_codes = pd.read_csv('Data/LSOA code to name 2019.csv')
del df_codes['FID']
```

```python
df_codes.rename(columns = {'LSOA11CD': 'LSOA code'}, inplace = True)
df_codes.rename(columns = {'LSOA11NM': 'LSOA name'}, inplace = True)
code_to_bor = df_codes.copy()
code_to_bor['LSOA name'] = code_to_bor['LSOA name'].str.split(' ').str[0]
code_to_bor = code_to_bor[code_to_bor['LSOA name'].isin(boroughs)]
code_to_bor.rename(columns = {'LSOA name': 'Borough'}, inplace = True)
code_to_bor_filt = code_to_bor[code_to_bor['Borough'].isin(boroughs)]
df.rename(columns = {'FeatureCode': 'LSOA code'}, inplace = True)
df_w_bor = pd.merge(code_to_bor_filt, df, on = 'LSOA code')
```

```python
df_rank = df_w_bor[df_w_bor['Measurement'] == 'Rank']
# 'Decile ' has one space after last letter for some reason
df_decile = df_w_bor[df_w_bor['Measurement'] == 'Decile ']
df_score = df_w_bor[df_w_bor['Measurement'] == 'Score']
```

```python
imd_score = df_score[df_score['Indices of Deprivation'] == 'a. Index of Multiple Deprivation (IMD)'].reset_index()
imd_score.sort_values(by = 'Value', ascending = True, inplace = True)
```

```python
top_n = 50
top_LSOA = imd_score.head(top_n)

plt.figure(figsize=(10, 6))
plt.bar(top_LSOA['LSOA code'], top_LSOA['Value'], color='skyblue')
plt.title('Top LSOA by IMD Value')
plt.xlabel('LSOA code')
plt.ylabel('IMD Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
for borough in range(len(boroughs)):
    print(boroughs[borough], (top_LSOA['Borough'] == boroughs[borough]).sum())
```

```python
# Compute ranking for average LSOA IMD score per borough
imd_score_big = df_score[df_score['Indices of Deprivation'] == 'a. Index of Multiple Deprivation (IMD)']
imd_score = imd_score_big[['LSOA code', 'Value']]
imd_score_bor = pd.merge(imd_score, code_to_bor, on = 'LSOA code')

average_values = imd_score_bor.groupby('Borough')['Value'].mean().reset_index()
average_values.columns = ['Borough', 'Average']
average_values.sort_values(by = 'Average', ascending = False, inplace = True)

plt.figure(figsize=(10, 6))
plt.bar(average_values['Borough'], average_values['Average'], color='skyblue')
plt.title('Average LSOA IMD Score per Borough')
plt.xlabel('Borough')
plt.ylabel('Average LSOA IMD Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
