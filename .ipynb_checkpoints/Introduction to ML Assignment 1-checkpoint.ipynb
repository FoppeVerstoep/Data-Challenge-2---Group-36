```python
#1 Loading the data
import csv
import numpy as np
import matplotlib.pyplot as mp
from matplotlib.ticker import MaxNLocator
import pandas as pd
import seaborn as sns
```

```python

buf = []
with open('turtle-2.csv') as file:
    turtles = csv.reader(file)
    for row in turtles:
        buf.append(row)
arr = np.array(buf)
#2 Defining x and y
y = arr[1:, 8]
x = arr[1:, :8]
xlabels = []
for i in arr[0, 0:8]:
    xlabels.append(i)
print(xlabels)
vallist = []
for i in range(0, len(xlabels)):
    xlist = []
    for j in x[:, i]:
        if xlabels[i] != 'Sex':
            xlist.append(float(j))
        else:
            xlist.append(j)
    vallist.append(xlist)
ylist = []
for i in y:
    ylist.append(int(i))
#3 Generating the scatterplots
for i in range(0, len(xlabels)):
    fig, ax = mp.subplots()
    arr_temp = np.array(vallist[i])
    if xlabels[i] == 'Sex' or 'weight' in vallist[i]:
        ax.scatter(vallist[i], y)
    else:
        ax.scatter([x * 10 for x in vallist[i]], y)
    if arr[0, i] != "Sex":
        ax.xaxis.set_major_locator(MaxNLocator(10))
        if 'weight' in xlabels[i]:
            ax.set_xlabel(f'{xlabels[i]} (g)')
        else:
            ax.set_xlabel(f'Shell {xlabels[i]} (mm)')
    else:
        ax.set_xlabel('Sex')
    fig.suptitle(f'Correlation between {xlabels[i]} and ring count')
    ax.set_ylabel('Number of rings')
```

```python
# 4: Correct labels
xlabels = xlabels[:8]
if 'Rings' not in xlabels:
    xlabels.append('Rings')
    vallist.append(ylist)
#5 Generate correlation values and basic statistics
mean = []
stdev = []
min_ = []
max_ = []
for i in range(1, len(xlabels)):
    mean.append(np.mean(vallist[i]))
    stdev.append(np.std(vallist[i]))
    min_.append(min(vallist[i]))
    max_.append(max(vallist[i]))
ass = {'mean': mean, 'standard dev': stdev, 'minimum': min_, 'maximum': max_}
df = pd.DataFrame(ass)
df = df.set_axis(xlabels[1:])

print(df)

arr_df = pd.DataFrame(arr[1:, 1:])

#6 Generate correlation heatmap
correlations = arr_df.corr()
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f", xticklabels = xlabels[1:], yticklabels = xlabels[1:])
mp.suptitle('Correlation Matrix between Turtle measurements')
mp.savefig('Heatmap')
```

```python
#5 Creating violin plot for Sex
violin_data = {'Sex': vallist[0], 'Rings': ylist}
ax = sns.violinplot(x = "Sex", y = "Rings", data = violin_data)
ax.set_ylabel('Number of rings')
mp.suptitle('Sex vs Number of Rings')
fig = ax.get_figure()
fig.savefig('violinplot')
```

```python
#6 Create X_int with dummy variable
data = pd.DataFrame(arr[1:, 0])
sex_encoded = np.array(pd.get_dummies(data), dtype = int)
X_int = np.concatenate((sex_encoded, arr[1:, 1:8].astype(float)), axis = 1)
r2_overall = []
all_max_models = []
print(X_int)
#7 Bin y
#train = 0.8, val = 0.1, test = 0.1
from sklearn.model_selection import train_test_split
y_bins = [i for i in range(0, 10)]
y_binned = np.digitize(ylist, y_bins)
#8 Split data into train, val, test
X_train, x_temp, y_train, y_temp = train_test_split(X_int, y_binned, train_size = 0.8)
X_val, X_test, y_val, y_test = train_test_split(x_temp, y_temp, train_size = (0.4/0.8))

#9 Linear regression
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(X_train, y_train)
y_pred = linear.predict(X_test)
r2 = linear.score(X_test, y_test)
print("R^2: ", r2)
mean_absolute_error = np.mean(abs(y_test - y_pred))
print("MAE: ", mean_absolute_error)
mean_squared_error = np.mean((y_test - y_pred)**2)
print("MSE: ", mean_squared_error)
mp.scatter(y_test, y_pred)
mp.suptitle('Actual vs predicted y-values')
mp.xlabel('Actual y-values')
mp.ylabel('Predicted y-values')
mp.savefig('Linearreg')
r2_overall.append(r2)
all_max_models.append((linear, r2, mean_absolute_error, mean_squared_error))
```

```python
#9 Polynomial regression
from sklearn.preprocessing import PolynomialFeatures
R2s = []
for i in range(2, 6):
    polyT = PolynomialFeatures(degree = i).fit_transform(X_train)
    polyV = PolynomialFeatures(degree = i).fit_transform(X_val)
    polyReg = LinearRegression()
    polyReg.fit(polyT, y_train)
    R2_val = polyReg.score(polyV, y_val)
    R2s.append((R2_val, polyReg))
    print(i)
max_r2 = max(R2s)
print(R2s)
print(max_r2)
values = [i[0] for i in R2s]
max_r2 = max(values)
max_index = [i for i in range(0, len(values)) if values[i] == max_r2][0]
print(f'Maximum R2: {max_r2} at degree {max_index + 2}')
polyR = PolynomialFeatures(degree = max_index + 2).fit_transform(X_test)
polyReg = R2s[max_index][1]
y_pred = polyReg.predict(polyR)
r2 = polyReg.score(polyR, y_test)
print(f'R2: {r2}')
mp.scatter(y_test, y_pred)
mp.suptitle('Actual vs predicted y-values')
mp.xlabel('Actual y-values')
mp.ylabel('Predicted y-values')
mp.savefig('polynomialreg')
mean_absolute_error = np.mean(abs(y_test - y_pred))
print("MAE: ", mean_absolute_error)
mean_squared_error = np.mean((y_test - y_pred)**2)
print("MSE: ", mean_squared_error)
r2_overall.append(r2)
all_max_models.append((polyR, max_index, r2, mean_absolute_error, mean_squared_error))
```

```python
#10 K-means regression
from sklearn.neighbors import KNeighborsRegressor
ks = {}
for i in(1, 3, 6, 10):
    
    mg = KNeighborsRegressor(n_neighbors = i)
    mg.fit(X_train, y_train)
    y_pred = mg.predict(X_val)
    score = mg.score(X_val, y_val)
    ks[i] = (mg, score)
scores = [i[1] for i in ks.values()]
models = [i[0] for i in ks.values()]
max_score = max(scores)
max_k = [i for i in ks.keys() if ks[i][1] == max_score]
print(f'validation score for best model: {max_score} at k {max_k[0]}')
mg = ks[max_k[0]][0]
y_pred = mg.predict(X_test)
r2 = mg.score(X_test, y_test)
print(f'R2: {r2}')
mp.scatter(y_test, y_pred)
mp.suptitle('Actual vs predicted y-values')
mp.xlabel('Actual y-values')
mp.ylabel('Predicted y-values')
mp.savefig('kmeans')
mean_absolute_error = np.mean(abs(y_test - y_pred))
print("MAE: ", mean_absolute_error)
mean_squared_error = np.mean((y_test - y_pred)**2)
print("MSE: ", mean_squared_error)
r2_overall.append(r2)
all_max_models.append((mg, max_k, r2, mean_absolute_error, mean_squared_error))
```

```python
#11 convert raw y to integers
y = np.array(ylist)
#12 Check among these options which one is the best:
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.inspection import permutation_importance
from sklearn.metrics import r2_score
model_list = [linear, polyReg, mg]
max_r2 = max(r2_overall)
max_val = [i for i in range(0, len(r2_overall)) if r2_overall[i] == max_r2][0]
if model_list[max_val] == linear:
    model = LinearRegression()
elif model_list[max_val] == polyReg:
    model = polynomialRegression(max_index[0] + 2)
elif model_list[max_val] == mg:
    model = KNeighborsRegressor(n_neighbors = max_k[0])
else:
    print('An error has occured in the registration of R-values.')
#12 Perform k-fold cross validation on the best model found.
cv = KFold(n_splits=10, random_state=None)
acc_score = []
X_split = cv.split(X_train.copy())
for train_index, test_index in X_split:
    X_train_val, X_test_val = X_int[train_index, :], X_int[test_index, :]
    y_train_val , y_test_val = y[train_index] , y[test_index]
    model.fit(X_train_val,y_train_val)
    y_pred = model.predict(X_test_val)
    acc = r2_score(y_test_val, y_pred)
    mse = np.mean((y_test_val - y_pred)**2)
    mae = np.mean(abs(y_test_val - y_pred))
    feature_variances = permutation_importance(model, X_test_val, y_test_val)
    acc_score.append((acc, mse, mae, feature_variances, model))
    

maes = [i[2] for i in acc_score]
mses = [i[1] for i in acc_score]
r2s = [i[0] for i in acc_score]
min_mae= (min(maes), [i for i in range(len(acc_score)) if maes[i] == min(maes)])
print(min_mae)
max_mae = (max(maes), [i for i in range(len(acc_score)) if maes[i] == max(maes)])
best_model = acc_score[min_mae[1][0]][3]
worst_model = acc_score[max_mae[1][0]][3]
mp.scatter([(i + 1) for i in range(len(r2s))],r2s)
mp.suptitle('R2s vs k-fold')
mp.xlabel('k-fold')
mp.ylabel('R^2')
mp.savefig('R2s vs k-fold')
mp.show()

mp.scatter([(i + 1) for i in range(len(r2s))],mses)
mp.suptitle('Mean Squared Errors vs k-fold')
mp.xlabel('k-fold')
mp.ylabel('MSE')
mp.savefig('mses vs k-fold')
mp.show()

mp.scatter([(i + 1) for i in range(len(r2s))],maes)
mp.suptitle('Mean Absolute Errors vs k-fold')
mp.xlabel('k-fold')
mp.ylabel('MAE')
mp.savefig('maes vs k-fold')
mp.show()

#12.1 Append data for best and worst models:
min_index = min_mae[1][0]
all_max_models.append((best_model, min_mae, mses[min_index], r2s[min_index]))
max_index = max_mae[1][0]
worst_final = (worst_model, max_mae, mses[max_index], r2s[max_index])
```
