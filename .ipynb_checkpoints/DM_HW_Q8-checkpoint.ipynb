```python
# get data
data = fetch_california_housing()
print(data['target'])
X = data['data']
y = data['target']
#print(data['DESCR'])
#print(type(data))

df = pd.DataFrame(data.data, columns=data.feature_names)
df['Target'] = data.target
#print(df.head())


#plot all features against target
# Define the y-axis columns for the scatter plots

y_columns = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'Latitude', 'Longitude']

# Fit and transform the data using Z-score scaling

df_scaled = df.copy()
scaler = StandardScaler()
df_scaled[y_columns] = scaler.fit_transform(df[y_columns])


# Create the figure and the subplots

fig, axes = plt.subplots(4, 2, figsize=(10, 15))

fig.suptitle('Scatter Plots of Scaled Features vs Target', fontsize=16)



# Plot each scatter plot in the figure

for idx, column in enumerate(y_columns):
    row = idx // 2
    col = idx % 2
    ax = axes[row, col]
    ax.scatter(df_scaled['Target'], df_scaled[column], alpha=0.7)
    ax.set_title(f'Scaled {column} vs Target')
    ax.set_xlabel('Target')
    ax.set_ylabel(f"Scaled {column}")

# Adjust layout

plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()




```
