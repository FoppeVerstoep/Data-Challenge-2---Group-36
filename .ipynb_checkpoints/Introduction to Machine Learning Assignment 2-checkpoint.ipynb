```python
#1 Loading the data
import csv
import numpy as np
import matplotlib.pyplot as mp
from matplotlib.ticker import MaxNLocator
import pandas as pd
import seaborn as sns
buf = []
with open('bots_vs_users.csv') as file:
    turtles = csv.reader(file)
    for row in turtles:
        buf.append(row)
df = pd.DataFrame(buf[1:], columns = buf[0])
keys = df.keys()
    
#2 Detecting NaN values, deleting collumns with too many NaN values
key_null_before = {}
del_keys = []
for i in keys:
    empty_count = 0
    total_count = 0
    collumn = df.loc[:, i]
    for j in collumn:
        total_count += 1
        if j == '' or j.lower() == 'unknown':
            empty_count += 1
    empty_ratio = empty_count / total_count
    key_null_before[i] = empty_ratio
    print(f'{i} & {empty_ratio} \\\\')
df_clean = df.copy()
for i in key_null_before.keys():
    if key_null_before[i] >= 0.7:
        print(i)
        df_clean = df_clean.drop(i, axis = 1)
        del_keys.append(i)
assert len(df_clean.keys()) != len(df.keys())
for i in del_keys:
    assert i not in df_clean.keys()
```

```python
display(df_clean)
# This is the dataset which is to be used with the understanding that there are null values. To confirm:
```

```python
#3 Printing list with before and after NULL value ratio
keys_clean = df_clean.keys()
key_null_after = {}
for i in keys_clean:
    empty_count = 0
    total_count = 0
    collumn = df_clean.loc[:, i]
    for j in collumn:
        total_count += 1
        if j == '' or j.lower() == 'unknown':
            empty_count += 1
    empty_ratio = empty_count / total_count
    key_null_after[i] = empty_ratio

print(key_null_before.keys())
print(key_null_after.keys())
for i in key_null_before.keys():
    if i in key_null_after.keys():
        print(f'{i} & {key_null_before[i]} & Kept \\\\')
    else:
        print(f'{i} & {key_null_before[i]} & Deleted\\\\')
df_clean_r = df_clean.replace('Unknown', None)
```

```python
one_hot_encoded = pd.get_dummies(df_clean['city'], prefix='city')

# Concatenate the one-hot encoded columns back to the original DataFrame
df_clean_r = pd.concat([df_clean_r, one_hot_encoded], axis=1)

# Drop the original 'Feature1' column if you no longer need it
df_clean_r = df_clean_r.drop('city', axis=1)
```

```python
#4 Creating X and Y lists, train_test splitting
from sklearn.model_selection import train_test_split
y = df_clean_r.target
X = df_clean_r.copy().drop('target', axis = 1)
assert 'target' not in X.keys()
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size = 0.2)
assert len(X_train) == round(0.8*len(X))
assert len(y_train) == round(0.8*len(y))
```

```python
#5 Decision Tree Model
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
print(clf.score(X_train, y_train))
y_pred = clf.predict(X_test)
confusion = confusion_matrix(y_test, y_pred)
print(confusion)
```
