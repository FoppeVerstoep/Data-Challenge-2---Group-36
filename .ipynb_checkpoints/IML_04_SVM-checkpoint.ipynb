# Support Vector Machines 

The support vector machines (SVM) algorithm finds a hyperplane (or decision boundary) which should ideally have the following properties:
* It separates between data of two classes with a maximum margin
* Its equation (w.x + b = 0) yields a value ≥ 1 for examples from +ve class and ≤-1 for examples from -ve class

# 1. Linear SVMs
Let's apply _SVC_ with linear models on the forge dataset. 

```python
from sklearn.svm import SVC
import mglearn
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs

X, y = mglearn.datasets.make_forge()

clf = SVC(kernel='linear', C=0.1, tol = 0.00001).fit(X, y)
plt.figure(1)
plt.scatter(X[:, 0], X[:, 1], c=y)
# we can plot the decision boundary and margins  
w = clf.coef_[0] 
a = -w[0] / w[1]
xx = np.linspace(np.amin(X[:, 0]),np.amax(X[:, 0])) 
yy = a*xx - clf.intercept_[0] / w[1]
plt.plot(xx,yy)
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, linewidth=1, facecolors='none', edgecolors='k')   

print(clf.coef_[0],clf.intercept_[0])
```

For  _SVC_ with linear kernels, the trade-off parameter that determines the strength of the regularization is called C. For higher values of the parameter C,  _SVC_ will try to fit the training set as best as possible, while for lower values of C, the models will put more emphasis on finding coefficient vector (w) that is close to zero.
```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
X_trainvalid, X_test, y_trainvalid, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_trainvalid, y_trainvalid, stratify=y_trainvalid, random_state=42)
```

```python
#
# Your code goes here
r2s = []
models = []
mses = []
from sklearn.metrics import r2_score
for i in (0.01, 1, 100):
    model = SVC(kernel='linear', C=i, tol = 0.00001).fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    r2s.append(model.score(X_valid, y_valid))
    mses.append(np.mean((y_valid - y_pred)**2))
    models.append(model)
print(r2s)
print(mses)
best_mse = (min(mses), [i for i in range(len(mses)) if mses[i] == min(mses)][0])
best_model = models[best_mse[1]]
test_r2 = best_model.score(X_test, y_test)
y_pred = best_model.predict(X_test)
test_mse = np.mean((y_test - y_pred)**2)
coefficients = best_model.coef_
plt.bar(range(len(coefficients[0])), coefficients[0])
plt.xlabel('Feature')
plt.ylabel('Coefficients')
plt.suptitle('Coefficients for different Features in the SVM')
print(f'Best model has C of {[0.01, 1, 100][best_mse[1]]} with mse = {[best_mse]} on the training data, mse = {test_mse} on the test data and test R2 of {test_r2}')
#
```

# 3. SVM for multiclass classification
Multiclass classification with linear SVMs can be done either through 
* the one-vs-rest scheme (multi_class = 'ovr'): in which for each class, a binary classification problem of data belonging or not to that class is done, or 
* the Crammer Singer scheme (multi_class = 'crammer_singer'): which optimizes a joint objective over all classes. 
While crammer_singer is interesting from a theoretical perspective as it is consistent, it is seldom used in practice as it rarely leads to better accuracy and is more expensive to compute. 

Let’s apply the crammer_singer method to a simple four-class classification dataset. We use a two-dimensional dataset, where each class is given by data sampled from a Gaussian distribution

```python
from sklearn.datasets import make_blobs
import mglearn
import matplotlib.pyplot as plt
import numpy as np

colors = ['red','green','blue','purple']

X, y = make_blobs(centers=4, random_state=8)
# converts blobs with label 3 and 1 to 1 
# converts blobs with labels 2 and 0 to 0 
#y = y % 2 
print(X.shape)
plt.scatter(X[:, 0], X[:, 1], c = y) 
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

The following code train a Linear SVM classifier on this dataset (with the Crammer Singer scheme) and plots the decision boundary. 

```python
from sklearn.svm import LinearSVC


linear_svm = LinearSVC(multi_class ='crammer_singer',max_iter=10000).fit(X, y)
print("Coefficient shape: ", linear_svm.coef_.shape)
print("Intercept shape: ", linear_svm.intercept_.shape)
print(linear_svm.coef_)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_, ['b', 'r', 'g','y']):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)


plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(['Class 0', 'Class 1', 'Class 2','Class 3', 'Line class 0', 'Line class 1','Line class 2','Line class 3'], loc=(1.01, 0.3))
```

# 4. Linear SVM models and nonlinear features

Kernelized support vector machines (often just referred to as SVMs) are an extension of linear SVMs that allows for more 
complex models. 

Linear models can be quite limiting in low-dimensional spaces, as lines and hyperplanes have limited flexibility. One way to make a linear model more flexible is by adding more features—for example, by adding interactions or polynomials of the input features. Let's look at this synthetic dataset. We create 4 centers, which will have 4 different labels $[0-3]$. The code $y = y\%2$ coverts blobs with label "3" and "1" to "1" and blobs with labels "2" and "0" to "0".

A linear model for classification can only separate points using a line, and will not be able to do a very good job on this dataset. Now let’s expand the set of input features, say by also adding $feature1^2$, the square of the second feature, as a new feature. Instead of representing each data point as a two-dimensional point, ($feature0$, $feature1$), we now represent it as a three dimensional point, ($feature0$, $feature1$, $feature1^2$).

```python
from mpl_toolkits.mplot3d import Axes3D

# converts blobs with label 3 and 1 to 1 
# converts blobs with labels 2 and 0 to 0 
y = y % 2
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c = y) 
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")

# add the squared  feature 1
X2 = X[:, 1:] ** 2


fig = plt.figure(figsize=(8, 8))
# visualize in 3D
ax = fig.add_subplot(projection='3d')
# plot first all the points with y == 0, then all with y == 1
mask = y == 0
ax.scatter(X[mask, 0], X[mask, 1], X2[mask, 0], c='b', cmap=mglearn.cm2, s=60)
ax.scatter(X[~mask, 0], X[~mask, 1], X2[~mask, 0], c='r', marker='^', cmap=mglearn.cm2, s=60)
ax.set_xlabel("feature0")
ax.set_ylabel("feature1")
ax.set_zlabel("feature1 ** 2")
```

In the new representation of the data, it is now indeed possible to separate the two
classes using a linear model, a plane in three dimensions. We can confirm this by fitting
a linear model to the augmented data

```python
X_new = np.hstack([X,X2]); 
linear_svm_3d = SVC(kernel = 'linear').fit(X_new, y)
coef, intercept = linear_svm_3d.coef_.ravel(), linear_svm_3d.intercept_

%matplotlib inline
print(linear_svm_3d.coef_[0])
# show linear decision boundary
figure = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')
xx = np.linspace(X_new[:, 0].min() - 2, X_new[:, 0].max() + 2, 50)
yy = np.linspace(X_new[:, 1].min() - 2, X_new[:, 1].max() + 2, 50)
XX, YY = np.meshgrid(xx, yy)
ZZ = (coef[0] * XX + coef[1] * YY + intercept)/ -coef[2]
ax.plot_surface(XX, YY, ZZ, rstride=8, cstride=8, alpha=0.3)
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b',cmap=mglearn.cm2, s=60)
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', marker='^',cmap=mglearn.cm2, s=60)
ax.set_xlabel("feature0")
ax.set_ylabel("feature1")
ax.set_zlabel("feature0 ** 2")
```

```python
#As a function of the original features, the linear SVM model is not actually linear anymore.
#It is not a line, but more of an ellipse, as you can see from the plot created here

ZZ = YY ** 2
dec = linear_svm_3d.decision_function(np.c_[XX.ravel(), YY.ravel(), ZZ.ravel()])

figure = plt.figure()
plt.contourf(XX, YY, dec.reshape(XX.shape), levels=[dec.min(), 0, dec.max()], cmap=mglearn.cm2, alpha=0.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
X_new = np.hstack([X,X2]); 
linear_svm_3d = SVC(kernel = 'linear').fit(X_new, y)
coef, intercept = linear_svm_3d.coef_.ravel(), linear_svm_3d.intercept_

print(linear_svm_3d.coef_[0])
# show linear decision boundary
figure = plt.figure()
ax = fig.add_subplot(projection='3d')
xx = np.linspace(X_new[:, 0].min() - 2, X_new[:, 0].max() + 2, 50)
yy = np.linspace(X_new[:, 1].min() - 2, X_new[:, 1].max() + 2, 50)
XX, YY = np.meshgrid(xx, yy)
ZZ = (coef[0] * XX + coef[1] * YY + intercept) / -coef[2]
ax.plot_surface(XX, YY, ZZ, rstride=8, cstride=8, alpha=0.3)
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b',
cmap=mglearn.cm2, s=60)
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', marker='^',cmap=mglearn.cm2, s=60)
ax.set_xlabel("feature0")
ax.set_ylabel("feature1")
ax.set_zlabel("feature0 ** 2")
```

# 5. Kernel SVMs
By adding nonlinear features to the representation of our data can make linear models much more powerful. However, often we don’t know which features to add, and adding many features (like all possible interactions in a 100-dimensional feature space) might make computation very expensive. 
Kernevl SVMs allow us to learn a classifier in a higher-dimensional space without actually computing the new, possibly very large representation. Let's train kernel SVMs on the following dataset with a Gaussian radial basis kernel. 

```python
from sklearn.svm import SVC
svm = SVC(kernel='rbf', C=10, gamma=0.1).fit(X, y)
mglearn.plots.plot_2d_separator(svm, X, eps=.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
# plot support vectors
sv = svm.support_vectors_
# class labels of support vectors are given by the sign of the dual coefficients
sv_labels = svm.dual_coef_.ravel() > 0
mglearn.discrete_scatter(sv[:, 0], sv[:, 1], sv_labels, s=15, markeredgewidth=3)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

The decision boundary is shown in black, and the support vectors are larger points with the wide outline. 
The gamma parameter controls the width of the Gaussian kernel. It determines the scale of what it means for points to be close together. A small gamma means a large radius for the Gaussian kernel, which means that many points are considered close by. A
low value of gamma means that the decision boundary will vary slowly, which yields a model of low complexity, while a high value of gamma yields a more complex model.
The C parameter is a regularization parameter, similar to that used in the linear models. It limits the importance of each point. A small C means a very restricted model, where each data point can only have very limited influence. You can see that at Increasing C allows these points to have a stronger influence on the model and makes the decision boundary bend to correctly
classify them.

# 6. SVM Regression 
As we mentioned earlier, the SVM algorithm is quite versatile: not only does it support linear and
nonlinear classification, but it also supports linear and nonlinear regression. The trick is to reverse the
objective: instead of trying to fit the largest possible street between two classes while limiting margin
violations. Let's load the following dataset to fit regression models. 

```python
from sklearn.svm import SVR
# #############################################################################
# Generate sample data
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()

# #############################################################################
# Add noise to targets
y[::5] += 3 * (0.5 - np.random.rand(8))

plt.figure()
plt.plot(X,y,'o')
plt.xlabel('X')
plt.ylabel('y')
```

The following code fits an rbf regression model and predicts the regression line. 

```python
svr_rbf = SVR(kernel='rbf', C=0.1, gamma=0.1, epsilon=.1)
y_rbf = svr_rbf.fit(X, y).predict(X)

plt.figure()
plt.plot(X,y,'o')
plt.plot(X,y_rbf)
plt.scatter(X[svr_rbf.support_],y[svr_rbf.support_], facecolor="none", edgecolor = 'r', s=100)
plt.xlabel('X')
plt.ylabel('y')
```

# 7. Scaling data for SVMs 
While SVMs often perform quite well, they are very sensitive to the settings of the parameters and to the scaling of the data. Scaling data is multiplying all instances of a variable by a constant to change the variable's range. You might want to scale your data so that 
1. your data are in the same unit types (e.g. you might have some measurements in kilograms and others in pounds so you will need to scale our data to the same unit type) 
2. if there are variables with huge differences in ranges, the one with the larger range may dominatore your prediction model even though it may be less important than variables with smaller ranges.  

Now let’s go back to the cancer dataset and see the effect of scaling on different classifiers. First, we will load the dataset and display the range of each feature on the log scale.  

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
plt.plot(X_train.min(axis=0), 'o', label="min")
plt.plot(X_train.max(axis=0), '^', label="max")
plt.legend(loc=4)
plt.xlabel("Feature index")
plt.ylabel("Feature magnitude")
plt.yscale("log")
```

From the previous plot, we can see that the features in the dataset are of completely different orders of magnitude. This can be somewhat of a problem for other models (like linear models), but it has devastating effects for the kernel SVM. One way to resolve this problem is by rescaling each feature so that they are all approximately on the same scale. A common rescaling method for kernel SVMs is to scale the data such that all features are between 0 and 1.

You can scale the dataset by subtract X_train by the min and divide by the range. The function _MinMaxScaler_ performs that operation for you as shown below. 

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
# transform training data
X_train_scaled = scaler.transform(X_train)
# print dataset properties before and after scaling
print("transformed shape: {}".format(X_train_scaled.shape))
print("per-feature minimum before scaling:\n {}".format(X_train.min(axis=0)))
print("per-feature maximum before scaling:\n {}".format(X_train.max(axis=0)))
print("per-feature minimum after scaling:\n {}".format(X_train_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(X_train_scaled.max(axis=0)))
```
